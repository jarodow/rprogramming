### Question 1

What was the average trip duration in each city?

```r
#Use summary to double check that the mean is calulated properly
summary(ny$Trip.Duration)
```

```r
# saving the mean to a variable
nymean <- mean(x=ny$Trip.Duration, na.rm=TRUE)
nymean
```

```r
#Use summary to double check that the mean is calulated properly
summary(chi$Trip.Duration)
```

```r
# saving the mean to a variable
chimean<- mean(x=chi$Trip.Duration, na.rm=TRUE)
chimean
```

```r
#Use summary to double check that the mean is calulated properly
summary(wash$Trip.Duration, na.rm=TRUE)
```

```r
# saving the mean to a variable
washmean <- mean(x=wash$Trip.Duration, na.rm=TRUE)
washmean
```

```r
#creating a dataframe of city means
dfcitymeans <- data.frame(city=c("New York", "Washington", "Chicago"),
                citymeans=c(nymean, washmean, chimean))
head(dfcitymeans)
```

```r
#creating a bar chart using ggplot for the means of trip duration per city
p<-ggplot(data=dfcitymeans, aes(x=city, y=citymeans)) +
  #filled the bar color to give it some color
  geom_bar(stat="identity", fill=("#990969")) +
  #added labels to the x and y axis
  labs(x="City", y="Average of Trip Duration") +
  #added a title
  ggtitle("Bar chart of average trip duration per city")
p
```

 It seems like people in Washington had a higher trip duration as compared to Chicago and New York. New York and Chicago seemed to have similar trip durations. I felt like a bar chart was the easiest way to compare the means of each cities trip duration.

### Question 2

How did the birth year effect trip duration? I also did a chart for the User counts by birth year.

```r
#Since only NY and CHI have birth year
#added both into a single dataframe
#added ny and chicago 
all<-rbind(ny,chi)
```

```r
ggplot(aes(x=all$Birth.Year, y=all$Trip.Duration),data=all) +
  #used a point graph to fit most of the data points gave it a low alpha to see the shape
  #also gave it a red color
  geom_point(alpha=(0.05),color='red')+
  #added labels and a title
  labs(x="Birth Year",y="Duration of Trip",title="New York & Chicago Duration of Trip by Birth Year")+
  #cut down the cartesian coordinates to better show the data
  coord_cartesian(ylim=c(0,5000),xlim=c(1925,2000))+
  #showed the mean of the data in yellow 
  geom_line(stat='summary',fun.y="mean", color = 'yellow')+
  #showed the lower .1 quantile
  geom_line(stat = 'summary', fun.y = "quantile", fun.args = list(probs = .1), color = 'blue') +
  #showed the top .9 quantile
  geom_line(stat = 'summary', fun.y = "quantile", fun.args = list(probs = .9), color = 'blue') +
  #showed the .5 quantile 
  geom_line(stat = 'summary', fun.y = "quantile", fun.args = list(probs = .5), color = 'green') 
```

```r
#User counts by Birth year
ggplot(aes(x=all$Birth.Year),data=all) +
  #added some color
  geom_histogram(fill="#040199")+
  #added some labels and a title
  labs(x="Birth Year",y="User Counts",title="New York & Chicago User Counts by Birth Year")+
  #cut down the coordinates to better show the data
  coord_cartesian(ylim=c(0,12500),xlim=c(1925,2000))
```

For the first chart I felt that a point chart was the best choice to easily show all the data points. I added the mean, median, and lower 10% and top 90% lines to make it a little easier to see the effect. I found that there wasn't that much of an effect on trip duration depending on age. I also decided to check the user counts by birth year. This showed that there was a much higher amount of people in the 1985 to 1990 range. This was probably why my first chart showed so many points around that age group. I felt the mean and quantile lines were necessary because of this. 

### Question 3

What days do most people start their trips?

```r
#I created a function to get the weekday of each city
getweekdaydf = function(val=0, name="") {
   #this stores the starttime as a weekday
   z<-weekdays(as.Date(val$Start.Time)) 
   #this adds the weekdays to a table
   k<-table(z)
    #this converts the table back to a dataframe
   weekdaydf<-as.data.frame(k)
    #this adjust the column names
   colnames(weekdaydf)<-c("Day","Number.Of.Start")
    #this puts the name of the city in a new column
   weekdaydf['City']<-c(name,name,name,name,name,name,name)
    #this line returns the newly made dataframe
   return(weekdaydf)
}
```

```r
#using the new function I make a weekday dataframe for Chicago
j<-getweekdaydf(val=chi,name="Chicago")
```

```r
#using the new function I make a weekday dataframe for Washington
k<-getweekdaydf(val=wash,name="Washington")
```

```r
#using the new function I make a weekday dataframe for New York
i<-getweekdaydf(val=ny,name="New York")
```

```r
#binding the dataframe together
h<-rbind(j,k)
```

```r
#binding the other dataframe together
l<-rbind(h,i)
```

```r
#setting the levels so the chart shows up properly
l$Day <-factor(l$Day, levels = c("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"))
```

```r
ggplot(l, aes(x = l$Day, y=l$Number.Of.Start, fill = l$City)) +
    #Set type of chart, Hid legend since showing the cities at the top
    geom_bar(stat="identity", show.legend=FALSE) +
    #Set the labels for the chart
    labs(x="Day", y="Number of uses", title="Uses per day in each City")+
    #Set the colors for each city
    scale_fill_manual(values = c("#0ADAEB", "#0A51A3","#550012")) + 
    #Set the facet grid to city to split by city 
    #felt like it made it easier to see the difference for each city 
    #since they had such large differences overall.
    facet_grid(.~City) +
    #Set the days to a horizontal style so they were readable
    theme(axis.text.x = element_text(angle=90, size=8)) +
    #Set the y breaks to make it a easier to see what number each one is at
    scale_y_continuous(breaks=seq(0,15000,2500))
```

For this last question I found that it varied a little bit for each city. Chicago was pretty similar each day with most people starting on Tuesdays or Fridays. For New York and Washington, Wednesdays were the busiest with it a drastic reduction on the weekends. It would probably be easiest to get a bike in New York and Washington during the weekend.
